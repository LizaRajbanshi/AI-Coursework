Task 6: HDI Category Column

```python
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)
```

```python
# HDI Category Distribution (Bar Chart)
category_counts = hdi_2022_df['HDI Category'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel("HDI Category")
plt.ylabel("Number of Countries")
plt.title("Distribution of Countries by HDI Category (2022)")
plt.tight_layout()
plt.show()
```

Task 2.3: Outlier Detection (IQR Method)

```python
#IQR Calculation
Q1_hdi = south_asia_2022['hdi'].quantile(0.25)
Q3_hdi = south_asia_2022['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

Q1_gni = south_asia_2022['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_2022['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni
```

```python
#Identify Outliers
outliers = south_asia_2022[
    (south_asia_2022['hdi'] < Q1_hdi - 1.5 * IQR_hdi) |
    (south_asia_2022['hdi'] > Q3_hdi + 1.5 * IQR_hdi) |
    (south_asia_2022['gross_inc_percap'] < Q1_gni - 1.5 * IQR_gni) |
    (south_asia_2022['gross_inc_percap'] > Q3_gni + 1.5 * IQR_gni)
]

outliers[['country', 'hdi', 'gross_inc_percap']]

```

```python
#Scatter Plot (Highlight Outliers)
plt.figure(figsize=(6,5))
plt.scatter(
    south_asia_2022['gross_inc_percap'],
    south_asia_2022['hdi'],
    label='Normal',
    alpha=0.7
)

plt.scatter(
    outliers['gross_inc_percap'],
    outliers['hdi'],
    color='red',
    label='Outliers'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita (South Asia, 2022)")
plt.legend()
plt.tight_layout()
plt.show()
```

Task 3.7: Outlier Detection (IQR)

```python
# IQR Method
def detect_outliers(df):
    Q1_hdi = df['hdi'].quantile(0.25)
    Q3_hdi = df['hdi'].quantile(0.75)
    IQR_hdi = Q3_hdi - Q1_hdi

    Q1_gni = df['gross_inc_percap'].quantile(0.25)
    Q3_gni = df['gross_inc_percap'].quantile(0.75)
    IQR_gni = Q3_gni - Q1_gni

    return df[
        (df['hdi'] < Q1_hdi - 1.5 * IQR_hdi) |
        (df['hdi'] > Q3_hdi + 1.5 * IQR_hdi) |
        (df['gross_inc_percap'] < Q1_gni - 1.5 * IQR_gni) |
        (df['gross_inc_percap'] > Q3_gni + 1.5 * IQR_gni)
    ]
```

Scatter Plot – Outliers

```python
sa_outliers = detect_outliers(south_asia_2020_22)

plt.figure(figsize=(6,5))

# Normal data points
plt.scatter(
    south_asia_2020_22['gross_inc_percap'],
    south_asia_2020_22['hdi'],
    alpha=0.6,
    label='Normal data'
)

# Outliers in red
plt.scatter(
    sa_outliers['gross_inc_percap'],
    sa_outliers['hdi'],
    color='red',
    s=60,
    label='Outliers'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("South Asia Outliers (2020–2022)")
plt.legend()
plt.tight_layout()
plt.show()

```
